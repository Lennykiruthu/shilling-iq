```python
import pandas as pd 
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
```

```python
nse_2007_df = pd.read_csv("/storage/local-git-repos/shilling-iq/Nairobi Securities Exchange All Stocks Prices 2007-2012/NSE_data_all_stocks_2007.csv")
```
